```python
# directed
def creatematrix(n):
    m = []
    for i in range(n):
        c = []
        for j in range(n):
            c.append(0)
        m.append(c)
    return m

def printmatrix(m):
    for i in m:
        print(i)

# Open the file and read the graph
with open("graph.txt", "r") as f:
    line = f.readline()
    l = line.split()
    n = int(l[0])  # Number of nodes
    m = int(l[1])  # Number of edges

    # Initialize adjacency matrix
    adj_matrix = creatematrix(n)

    # Populate adjacency matrix for directed graph
    for i in range(m):
        line = f.readline()
        u = int(line.split()[0])
        v = int(line.split()[1])
        adj_matrix[u][v] = 1

# Print adjacency matrix (optional)
print("Adjacency Matrix:")
printmatrix(adj_matrix)
print("\n")

# Calculate out-degree (count 1's in each row)
out_degrees = [0] * n
for i in range(n):
    out_degrees[i] = sum(adj_matrix[i])

# Calculate in-degree (count 1's in each column)
in_degrees = [0] * n
for j in range(n):
    in_degrees[j] = sum(row[j] for row in adj_matrix)

# Print out-degree and in-degree for each node
for i in range(n):
    print(f"Node {i}: Out-Degree {out_degrees[i]}, In-Degree {in_degrees[i]}")
```

```python
# directed and weighted
def creatematrix(n):
    m = []
    for i in range(n):
        c = []
        for j in range(n):
            c.append(0)
        m.append(c)
    return m

def printmatrix(m):
    for i in m:
        print(i)

# Open the file and read the graph
with open("weightedgraph.txt", "r") as f:
    line = f.readline()
    l = line.split()
    n = int(l[0])  # Number of nodes
    m = int(l[1])  # Number of edges

    # Initialize adjacency matrix
    adj_matrix = creatematrix(n)

    # Populate adjacency matrix for directed weighted graph
    for i in range(m):
        line = f.readline()
        u = int(line.split()[0])
        v = int(line.split()[1])
        w = int(line.split()[2])
        adj_matrix[u][v] = w  # Use weight as the value in the adjacency matrix

# Print the adjacency matrix (optional)
print("Adjacency Matrix:")
printmatrix(adj_matrix)
print("\n")

# Calculate out-degree (count non-zero values in each row)
out_degrees = [0] * n
for i in range(n):
    out_degrees[i] = sum(1 for x in adj_matrix[i] if x != 0)

# Calculate in-degree (count non-zero values in each column)
in_degrees = [0] * n
for j in range(n):
    in_degrees[j] = sum(1 for row in adj_matrix if row[j] != 0)

# Print out-degree and in-degree for each node
for i in range(n):
    print(f"Node {i}: Out-Degree {out_degrees[i]}, In-Degree {in_degrees[i]}")
```

```python
# directed
def createlsit(n):
    g = {}
    for i in range(n):
        g[i] = []
    return g

def printlist(g):
    for i in range(len(g)):
        print(i, end=": ")
        print(g[i])

# Open the file and read the graph
f = open("graph.txt", "r")
line = f.readline()
l = line.split()
n = int(l[0])  # Number of nodes
m = int(l[1])  # Number of edges

# Initialize adjacency list
adj_list = createlsit(n)

# Populate adjacency list for directed graph
for i in range(m):
    line = f.readline()
    u = int(line.split()[0])
    v = int(line.split()[1])
    adj_list[u].append(v)  # Add directed edge from u to v

# Close the file after reading
f.close()

# Print the adjacency list
print("Adjacency List:")
printlist(adj_list)
print("\n")

# Calculate and print out-degree and in-degree for each node
out_degrees = [0] * n  # Initialize out-degree list
in_degrees = [0] * n   # Initialize in-degree list

# Calculate out-degree (length of the adjacency list)
for node in range(n):
    out_degrees[node] = len(adj_list[node])

# Calculate in-degree by checking each adjacency list
for node in range(n):
    for neighbor in adj_list[node]:
        in_degrees[neighbor] += 1  # If a node is a neighbor, increment its in-degree

# Print the out-degree and in-degree for each node
print("Out-Degree and In-Degree of each node:")
for node in range(n):
    print(f"Node {node}: Out-Degree {out_degrees[node]}, In-Degree {in_degrees[node]}")
```

```python
def createlsit(n):
    g = {}
    for i in range(n):
        g[i] = []
    return g

def printlist(g):
    for i in range(len(g)):
        print(i, end=": ")
        print(g[i])

# Open file and read the graph
f = open("weightedgraph.txt", "r")
line = f.readline()
l = line.split()
n = int(l[0])  # Number of nodes
m = int(l[1])  # Number of edges

# Initialize adjacency list
adj_list = createlsit(n)

# Populate adjacency list for directed weighted graph
for i in range(m):
    line = f.readline()
    u = int(line.split()[0])  # Source node
    v = int(line.split()[1])  # Destination node
    w = int(line.split()[2])  # Weight of the edge
    adj_list[u].append((v, w))  # Add directed edge from u to v with weight w

# Close the file after reading
f.close()

# Print the adjacency list
print("Adjacency List for Directed Weighted Graph:")
printlist(adj_list)

# Calculate out-degree (number of outgoing edges for each node)
out_degrees = [0] * n
for u in range(n):
    out_degrees[u] = len(adj_list[u])

# Calculate in-degree (number of incoming edges for each node)
in_degrees = [0] * n
for u in range(n):
    for v, w in adj_list[u]:
        in_degrees[v] += 1  # Node v receives an edge from u

# Print the out-degree and in-degree for each node
print("\nOut-Degree and In-Degree of each node:")
for i in range(n):
    print(f"Node {i}: Out-Degree {out_degrees[i]}, In-Degree {in_degrees[i]}")
```

```python
def createlist(n):
    g = {}
    for i in range(n):
        g[i] = []
    return g

def printlist(g):
    for i in range(len(g)):
        print(i, end=": ")
        print(g[i])

# Open file and read the graph
f = open("weightedgraph.txt", "r")
line = f.readline()
l = line.split()
n = int(l[0])  # Number of nodes
m = int(l[1])  # Number of edges

# Initialize adjacency list
adj_list = createlist(n)

# Populate adjacency list for directed weighted graph
for i in range(m):
    line = f.readline()
    u = int(line.split()[0])  # Source node
    v = int(line.split()[1])  # Destination node
    w = int(line.split()[2])  # Weight of the edge
    adj_list[u].append((v, w))  # Add directed edge from u to v with weight w

# Close the file after reading
f.close()

# Print the adjacency list
print("Adjacency List for Directed Weighted Graph:")
printlist(adj_list)

# Initialize lists for degrees and weight sums
out_degrees = [0] * n
in_degrees = [0] * n
out_weights = [0] * n
in_weights = [0] * n

# Calculate out-degree and outgoing weights
for u in range(n):
    out_degrees[u] = len(adj_list[u])
    out_weights[u] = sum(w for v, w in adj_list[u])

# Calculate in-degree and incoming weights
for u in range(n):
    for v, w in adj_list[u]:
        in_degrees[v] += 1
        in_weights[v] += w

# Print the out-degree, in-degree, and weight sums for each node
print("\nOut-Degree, In-Degree, Outgoing Weight, and Incoming Weight of each node:")
for i in range(n):
    print(f"Node {i}: Out-Degree {out_degrees[i]}, In-Degree {in_degrees[i]}, Outgoing Weight {out_weights[i]}, Incoming Weight {in_weights[i]}")
```
